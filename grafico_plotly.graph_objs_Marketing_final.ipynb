##Gráfico de Dispersão Tridimensional - ROI em relação à Data e Custo

```python
import plotly.graph_objs as go

datas = df['data'].tolist()
custos = df['custo'].tolist()
rois = df['retorno_investimento'].tolist()

# Criação do gráfico de dispersão tridimensional
fig = go.Figure(data=[go.Scatter3d(
    x=datas,
    y=custos,
    z=rois,
    mode='markers',
    marker=dict(
        size=8,
        color=rois,  # Usando a coluna ROI para colorir os pontos
        colorscale='Viridis',  # Escolha da escala de cores
        opacity=0.8
    )
)])

# Adição de rótulos e título
fig.update_layout(
    scene=dict(
        xaxis=dict(title='Data'),
        yaxis=dict(title='Custo'),
        zaxis=dict(title='ROI'),
    ),
    title='Gráfico de Dispersão Tridimensional - ROI em relação à Data e Custo'
)

# Exibição do gráfico
fig.show()

'''
Este código cria um gráfico especial chamado "Gráfico de Dispersão Tridimensional"
que nos ajuda a visualizar três importantes métricas de marketing de uma só vez:
Data, Custo e Retorno sobre o Investimento (ROI). Vamos entender cada parte:

Data: No eixo x (horizontal) do gráfico, temos a linha do tempo representada
por datas específicas. Isso nos permite ver como o custo e o ROI variam ao
longo do tempo.

Custo: No eixo y (vertical) do gráfico, temos o custo associado a cada data.
O custo representa quanto dinheiro foi gasto em uma determinada campanha ou
atividade de marketing em um determinado período de tempo.

ROI (Retorno sobre o Investimento): No eixo z (profundidade) do gráfico, temos
o ROI associado a cada data e custo. O ROI é uma métrica crítica em marketing
que nos diz quanto retorno financeiro obtemos em relação ao custo que investimos
em uma determinada campanha ou atividade de marketing.

Interpretação do Gráfico:

Cada ponto no gráfico representa uma combinação única de data, custo e ROI.
A cor de cada ponto no gráfico indica o valor do ROI. Pontos em cores mais
claras indicam um ROI mais alto, enquanto pontos em cores mais escuras indicam
um ROI mais baixo.
Podemos observar como o custo e o ROI variam ao longo do tempo e identificar
 quais períodos resultaram em um ROI mais alto ou mais baixo em comparação
 com o custo associado.

Importância do Gráfico:

O gráfico nos ajuda a entender visualmente a eficácia das campanhas de marketing
 ao longo do tempo.
Ele nos permite identificar tendências e padrões que podem nos ajudar a otimizar
 futuras estratégias de marketing para obter um melhor ROI.
Em resumo, este gráfico tridimensional nos dá uma visão abrangente e visualmente
 atraente das relações entre data, custo e ROI em nossas atividades de marketing.
'''
```

## Público alvo e taxa de conversão

```python
import plotly.graph_objs as go
from plotly.colors import qualitative

# Agrupa os dados pelo público-alvo e calcula a contagem de registros para cada categoria
grupo_publico_alvo = df.groupby('publico_alvo').count().reset_index()

# Transforma a taxa de conversão em porcentagem
grupo_publico_alvo['taxa_conversao'] *= 100

# Ordena o dataframe pelo valor da taxa de conversão de forma descendente
grupo_publico_alvo = grupo_publico_alvo.sort_values(by='taxa_conversao', ascending=False)

# Define o layout do gráfico
layout = go.Layout(
    title='Soma da Taxa de Conversão por Público-Alvo',
    xaxis=dict(title='Público-Alvo'),
    yaxis=dict(title='Soma da Taxa de Conversão (%)'),  # Atualiza o rótulo do eixo Y
    height=600,
    colorway=qualitative.Pastel  # Cores em degradê de azul
)

# Cria o trace de linhas
trace = go.Scatter(
    x=grupo_publico_alvo['publico_alvo'],
    y=grupo_publico_alvo['taxa_conversao'],
    mode='lines+markers',  # Define que o trace terá linhas e marcadores
    line=dict(color='blue', width=2),  # Define a cor e a largura da linha
    marker=dict(color='blue', size=10),  # Define a cor e o tamanho dos marcadores
)

# Cria a figura
fig = go.Figure(data=[trace], layout=layout)

# Adiciona mais marcações à média da taxa de conversão
fig.update_yaxes(nticks=10)

# Exibe o gráfico
fig.show()
```

##Comparação entre Custo Médio e Custo Mediano

```python
import matplotlib.pyplot as plt

# Criar as variáveis para média e mediana do custo
media_custo = df['custo'].mean()
mediana_custo = df['custo'].median()

# Plotar o boxplot
plt.figure(figsize=(8, 6))
plt.boxplot([df['custo'], df['custo']], labels=['Custo', 'Custo'], showmeans=True)  # Adiciona as médias
plt.scatter([1, 2], [media_custo, mediana_custo], color='red', label='Média/Mediana')  # Marcadores para média e mediana
plt.xlabel('Variável')
plt.ylabel('Custo')
plt.title('Boxplot da Média e Mediana do Custo')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Criar as variáveis para média, mediana e desvio padrão do custo
media_custo = df['custo'].mean()
mediana_custo = df['custo'].median()
desvio_padrao_custo = df['custo'].std()

# Plotar o boxplot
plt.figure(figsize=(8, 6))
plt.boxplot([df['custo'], df['custo']], labels=['Custo', 'Custo'], showmeans=True)  # Adiciona as médias
plt.scatter([1, 2], [media_custo, mediana_custo], color='red', label='Média/Mediana')  # Marcadores para média e mediana

# Adicionar barras de erro para o desvio padrão
plt.errorbar([1, 2], [media_custo, mediana_custo], yerr=desvio_padrao_custo, fmt='o', color='blue', label='Desvio Padrão')

plt.xlabel('Variável')
plt.ylabel('Custo')
plt.title('Boxplot da Média, Mediana e Desvio Padrão do Custo')
plt.legend()
plt.show()
```
