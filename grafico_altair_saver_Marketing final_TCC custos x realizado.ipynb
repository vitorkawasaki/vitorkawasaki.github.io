## Teste de calculo de custos

```python
#teste de calculo de custos
caminho_planilha = ('/content/Marketing_final.csv')
df = pd.read_csv(caminho_planilha)
total_custos =df['custo'].sum()
print(total_custos)
```

```python
grupo_empresa = df.groupby(['empresa','segmento','canal']).agg({'custo': 'sum', 'retorno_investimento': 'mean', 'engajamento': 'mean', 'impressao': 'sum'})
grupo_empresa=grupo_empresa.round(2)
print(grupo_empresa)
```

>Em resumo, o código cria um gráfico de barras empilhadas que ajuda os profissionais de marketing a entender como o custo total de uma campanha de marketing é distribuído entre diferentes empresas e canais de distribuição, segmentados por diferentes segmentos de mercado. Isso permite identificar quais empresas e canais estão contribuindo mais ou menos para o custo total, ajudando a otimizar os gastos de marketing e maximizar o retorno sobre o investimento.

## Gráficos dos dados em Altair_saver

```python
# Transformar o índice multinível em colunas
grupo_empresa_reset = grupo_empresa.reset_index()

# Criar um gráfico Altair
chart = alt.Chart(grupo_empresa_reset).mark_bar().encode(
    x=alt.X('empresa:N', title='Empresa - Canal', axis=alt.Axis(labelAngle=45)),
    y=alt.Y('sum(custo):Q', title='Custo Total'),
    color=alt.Color('segmento:N', title='Segmento'),
    column=alt.Column('canal:N', title='Canal')
).properties(
    width=alt.Step(80),  # Largura das colunas
    title='Custo Total por Empresa e Canal'
)

# Exibir o gráfico
chart
```

```python
grupo_publico_alvo = df.groupby(['publico_alvo','canal']).agg({'retorno_investimento': 'mean', 'engajamento': 'mean', })
grupo_publico_alvo=grupo_publico_alvo.round(2)
print(grupo_publico_alvo)
```

>Em resumo, este código permite visualizar como o engajamento e o retorno de investimento variam entre diferentes públicos-alvo e canais de marketing, fornecendo insights valiosos para estratégias de marketing e tomada de decisões.

```python
# Agrupar os dados por público-alvo e canal
grupo_publico_alvo = df.groupby(['publico_alvo', 'canal']).agg({
    'retorno_investimento': 'mean',
    'engajamento': 'mean'
}).reset_index()

# Arredondar os valores
grupo_publico_alvo = grupo_publico_alvo.round(2)

# Criar o gráfico Altair
grafico_engajamento = alt.Chart(grupo_publico_alvo).mark_line().encode(
    x='canal',
    y='engajamento',
    color='publico_alvo:N'
).properties(
    width=600,
    height=300,
    title='Engajamento Médio por Público-Alvo e Canal'
)

grafico_retorno = alt.Chart(grupo_publico_alvo).mark_line().encode(
    x='canal',
    y='retorno_investimento',
    color='publico_alvo:N'
).properties(
    width=600,
    height=300,
    title='Retorno de Investimento Médio por Público-Alvo e Canal'
)

# Exibir os gráficos
grafico_engajamento | grafico_retorno
```
